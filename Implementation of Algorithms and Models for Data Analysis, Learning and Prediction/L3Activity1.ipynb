```r
library(xlsx)
```

```r
data<-read.xlsx("NHLTop100.xlsx",sheetName="Top 100")
```

```r
data<-data[c("Player","G","A")]
```

```r
#Fitting the initial model with assists on Y axis and goals on X axis 
model1 <- lm((A) ~ (G),data=data)
plot((data$G),(data$A),ylim=c(0,2500),xlim=c(0,1000))
abline(model1,col="red",lwd=2)
```

```r
# Evaluating R^2 and p values of the model
summary(model1)
# As we can see the R-squared values are:
# Multiple R-squared:  0.1082,	Adjusted R-squared:  0.09913 
# p-value: 0.0008317
```

```r
#Fitting the line through Player Wayne Gretzky
model2 <- (lm(I(A-1963)~I(G-894)+0, data = data))
plot((data$G),(data$A),ylim=c(0,2500),xlim=c(0,1000))
abline(predict(model2,newdata=list(G=0))+1963,coef(model2),col="red",lwd=2)
```

```r
# Evaluating R^2 and p values of the model2
summary(model2)
# As we can see the R-squared values are:
# Multiple R-squared:  0.9091, Adjusted R-squared:  0.9082
# p-value: < 2.2e-16
```

```r
#Adding a new player "Patrick Kane" with G=285 and A=465
ndf<-data.frame("Patrick Kane",285,465)
names(ndf)<-c("Player","G","A")
data<-rbind(data,ndf)
```

```r
#Fitting the line through Player Patrick Kane
model3 <- (lm(I(A-465)~I(G-285)+0, data = data))
plot((data$G),(data$A),ylim=c(0,2500),xlim=c(0,1000))
abline(predict(model3,newdata=list(G=0))+465,coef(model3),col="red",lwd=2)
```

```r
#Evaluating R^2 and p values of the model3
summary(model3)
# As we can see the R-squared values are:
# Multiple R-squared:  0.5732, Adjusted R-squared:  0.5689
# p-value: < 2.2e-16
```

```r
#Passing the line through origin and Wayne Gretzky we get slope=2.191 which is approximately equal to required slope 2.19
model4<- (lm(I(A+365)~I(G) + 0, data))
plot((data$G),(data$A),ylim=c(0,2500),xlim=c(0,1000))
abline(model4,col="red",lwd=2)
summary(model4)
```

```r
#Passing the line through origin and Patrick Kane we get slope=1.630 which is approximately equal to required slope 1.63
model5<- (lm(I(A+77)~I(G) + 0, data))
plot((data$G),(data$A),ylim=c(0,2500),xlim=c(0,1000))
abline(model5,col="red",lwd=2)
summary(model5)
```

```r
#Fitting the initial model with assists on Y axis and goals on X axis and passing it through origin
model6 <- lm((A) ~ (G)-1,data=data)
plot((data$G),(data$A),ylim=c(0,2500),xlim=c(0,1000))
abline(model6,col="red",lwd=2)
summary(model6)
```

```r
#Tabulating all the results with adjusted R^2
xdf1<-data.frame("Model1","0.09913","0.0008317")
names(xdf1)<-c("Model Name","Adjusted R^2 value","p-value")
xdf2<-data.frame("Model2","0.9082","< 2.2e-16")
names(xdf2)<-c("Model Name","Adjusted R^2 value","p-value")
xdf3<-data.frame("Model3","0.5689","< 2.2e-16")
names(xdf3)<-c("Model Name","Adjusted R^2 value","p-value")
xdf4<-data.frame("Model4","0.9216","< 2.2e-16")
names(xdf4)<-c("Model Name","Adjusted R^2 value","p-value")
xdf5<-data.frame("Model5","0.9058","< 2.2e-16")
names(xdf5)<-c("Model Name","Adjusted R^2 value","p-value")
xdf6<-data.frame("Model6"," 0.8975","< 2.2e-16")
names(xdf6)<-c("Model Name","Adjusted R^2 value","p-value")
xdf<-rbind(xdf1,xdf2,xdf3,xdf4,xdf5,xdf6)
```
