```r
library(FNN)
#Reading data into a df
data<-read.csv("german_credit_data1.csv")
#Removing NA values
data<-na.omit(data)
#Setting sampling rate
```

```r
#For 80%
sampling.rate<-0.8
#Total number of points
n.points<-nrow(data)
num.test.set.labels <- n.points * (1 - sampling.rate)
#randomly sample which rows will go in the training set
training <- sample(1:n.points, sampling.rate * n.points,replace=FALSE)
#Define the training set
train <- subset(data[training, ], select = c(Age,factor(Sex),factor(Job),factor(Housing),factor(Saving.accounts),factor(Checking.account),Credit.amount,Duration,factor(Purpose)))
#the other rows are going into the test set
testing <- setdiff(1:n.points, training)
#define the test set to be the other rows
test <- subset(data[testing, ], select = c(Age,factor(Sex),factor(Job),factor(Housing),factor(Saving.accounts),factor(Checking.account),Credit.amount,Duration,factor(Purpose)))
# this is the subset of labels for the training set
cl <- data$Credit.Risks[training]
# subset of labels for the test set, we're withholding these
true.labels <- data$Credit.Risks[testing]
```

```r
#Looping through 1:20 to check misclassification rate
y<-0
for (k in 1:20) {
predicted.labels <- knn(train, test, cl, k)
# We're using the R function knn()
num.incorrect.labels <- sum(predicted.labels != true.labels)
misclassification.rate <- num.incorrect.labels /
num.test.set.labels
x<-misclassification.rate
y<-c(y,x)   
}
df80<-data.frame(y)
df80<-df80[2:21,]
df80<-data.frame(df80)
colnames(df80)[1] <- "train=80"
```
